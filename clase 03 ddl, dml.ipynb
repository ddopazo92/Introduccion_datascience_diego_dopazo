```python
import pandas as pd
import sqlite3
```

```python
def execute_query(connection, query):
  cursor = connection.cursor()
  try:
    cursor.execute(query)
    connection.commit()
    print("Query ejecutado satisfactoriamente")
  except Error as e:
    print(f"El error '{e}' ha ocurrido ")
```

```python
path = '/content/Clase 3 - Redes.sqlite'

sql = sqlite3.connect(path)

cursor = sql.cursor()
```

```python
cursor.execute ("SELECT name from sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
query = 'select * from users'
cursor.execute(query)
print(cursor.fetchall())
```

```python
eliminar_tabla = """
DROP TABLE IF EXISTS usuarios;
"""
execute_query(sql, eliminar_tabla)
```

```python
crear_tabla_usuarios = """
CREATE TABLE Usuarios (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  nombre VARCHAR(30) NOT NULL,
  apellido VARCHAR(30) NOT NULL,
  edad INTEGER,
  genero CHAR,
  nacionalidad VARCHAR(30)
);
"""
execute_query(sql, crear_tabla_usuarios)
```

```python
crear_tabla_posteos = """
CREATE TABLE IF NOT EXISTS Posteos (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  titulo VARCHAR(30) NOT NULL,
  descripcion VARCHAR(30) NOT NULL,
  user_id INTEGE NOT NULL,
  foreign key (user_id) references Usuarios(id)
);
"""
execute_query(sql, crear_tabla_posteos)
```

```python
crear_tabla_comentarios = """
CREATE TABLE IF NOT EXISTS Comentarios (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  texto VARCHAR(30) NOT NULL,
  user_id INTEGER NOT NULL,
  posteo_id INTEGER NOT NULL,
  foreign key (user_id) references Usuarios(id),
  foreign key (posteo_id) references Posteos(id)
);
"""
execute_query(sql, crear_tabla_comentarios)
```

```python
crear_tabla_me_gusta = """
CREATE TABLE Me_Gusta (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  user_id INTEGER NOT NULL,
  posteo_id INTEGER NOT NULL,
  foreign key (user_id) references Usuarios(id)
  foreign key (posteo_id) references Posteos(id)
);
"""
execute_query(sql, crear_tabla_me_gusta)
```
