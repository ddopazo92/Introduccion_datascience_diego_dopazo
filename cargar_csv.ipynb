AGRUPACION DE DATOS   **group by**

```python
agrupar = df.groupby('Platform')['Year_of_Release'].mean() ## Agrupo primero por columna que va en (), y los valores de la otra columna que va con []
                                                           ## con el metodo mean() asociado al Year_of_release te da el año promedio con mas registros de esa consola(columa seleccionada)

agrupar
```
