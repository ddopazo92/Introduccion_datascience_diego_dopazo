```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
links = ['https://salones-es.lorealprofessionnel.com/uy/departamento-de-montevideo']

for i in range(2,9):
  links.append('https://salones-es.lorealprofessionnel.com/uy/departamento-de-montevideo?page='+str(i))
len(links)

##Codigo para generar paginas y poder utilizarlas sin tener que copiar una por una
```

```python
links
```

```python
direcciones = []
nombres = []
```

```python
for i in links:
  url = i
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  for j in soup.find_all('span',class_="lf-location-default__content__container__header__title__text lf-geo-divisions__top__wrapper__main__locations__list__location__content__container__header__title__text"):
        nombres.append(j.text)
  for k in soup.find_all('span',class_="lf-parts-address__content__street lf-location-default__content__container__address__content__street lf-geo-divisions__top__wrapper__main__locations__list__location__content__container__address__content__street"):
        direcciones.append(k.text)
```

```python
dic = {'Nombre':nombres,'Direccion':direcciones}
df = pd.DataFrame(dic)
df
```
